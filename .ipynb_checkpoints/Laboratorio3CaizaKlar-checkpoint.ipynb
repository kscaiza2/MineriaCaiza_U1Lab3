## Regresión Lineal

```python
'''Importamos los modulos necesarios '''
%matplotlib notebook
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
'''Se le asigno default_rng con el fin de obtener una nueva instancia
de un Generador'''
rng = np.random.default_rng()
'''Generador aleatorio de datos, enviamos por parametro para que genere
un total de 100 datos para X y para Y genera los datos y toma el 
resultado de la funcion'''
x = rng.random(1000)
y = 1.6*x + rng.random(1000)
```

```python
'''Calcula una regresion lineal mediante minimos cuadrados para dos conjuntos
de datos, devuelve distintos valores como la pendiente de la recta de 
regresion, coeficiente de correlacion y demas'''
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
'''Funcion auxiliar que utiliza la pendiente y la interseccion con el fin
de devolver un nuevo valor'''
def fitted_value(x):
 return slope * x + intercept
```

```python
'''Calcular los valores ajustados para el eje Y '''
new_values = list(map(fitted_value, x))
```

```python
'''Generamos el grafico de dispersion original'''
plt.plot(x, y, 'o', label='original data')
'''Generamos la grafica de la linea divisoria o dispersion original'''
plt.plot(x, new_values, 'r', label='fitted line')
plt.legend()
```
